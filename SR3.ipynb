```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
colors=['#FF69B4', '#33CCCC', '#6699CC']

data = np.random.normal(loc=20.0,scale=5.0,size=365)
plt.hist(data,bins=20,color=colors[1], edgecolor='black',alpha=0.8)
plt.title('Histogram of CountryName vs Year', fontsize=18, color=colors[1])
plt.xlabel('Country Name', fontsize=14, color=colors[1])
plt.ylabel('Year', fontsize=14, color=colors[1])
plt.grid(True)
plt.background_color='#F7F7F7'
plt.legend()
plt.show()
```

```python
country_name = ['America', 'Iraq', 'Turkey', 'South Africa']
Year=[2003,2002,2001,2000,]
colors=['#FF69B4', '#FFFFFF', '#454545','#33CC33']
```

```python
plt.figure(figsize=(7,7))
plt.pie(Year, labels=country_name,autopct='%1.1f%%',colors=colors,shadow=True,startangle=90)
plt.title('pieChart', loc='center', fontsize=12)
plt.background_color='#F7F7F7'
plt.grid(color='#CCCCCC', linestyle='--', linewidth=0.5)
plt.show()
```

```python
country_name=['South Africa', 'Turkey', 'Iraq', 'America']
Year=[2002,2001,2000,1999]

plt.figure(figsize=(8,5))
plt.bar(country_name,Year, color='#FF69B4', alpha=0.5)
plt.title('Country name vs Years', fontsize=18, color='#333333')
plt.xlabel('Country name', fontsize=14, color='#333333')
plt.ylabel('Years', fontsize=14, color='#333333')
plt.xticks(rotation=90)
plt.grid(color='#CCCCCC', linestyle='--', linewidth=0.5)
plt.legend(title='Country_name vs Year', loc='center', fontsize=14)
plt.show()
```

```python
data = np.random.normal(size=1000,loc=10.0, scale=5.0)
sns.kdeplot(data,shade=True, color='#FF69B4', label='Desinty plot' ,alpha=0.8)
sns.rugplot(data, color='#CCCCCC')


plt.figure(figsize=(12, 6))           
plt.title('Density Graph', fontsize=18)
fontweigth=('bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(color='#444444', linestyle='--', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
